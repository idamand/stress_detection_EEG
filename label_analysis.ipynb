```python
from utils.labels import get_pss_labels, get_stai_labels
from scipy.stats import ttest_ind
from utils.valid_recordings import get_valid_recordings
from utils.data_processing import dict_to_arr
import numpy as np
```

```python
valid_recordings = get_valid_recordings('raw')
stai_labels_dict = get_stai_labels(valid_recordings=valid_recordings)
pss_labels_dict = get_pss_labels(valid_recordings=valid_recordings)
```

```python
def dict_to_arr(data_dict):
    keys_list = list(data_dict.keys())
    
    data_arr = np.empty(len(keys_list))
    
    i = 0
    for key in keys_list:
        data = data_dict[key]
        data_arr[i] = data
        i += 1

    return data_arr
```

```python
stai_labels = dict_to_arr(stai_labels_dict)
stai_labels
```

```python
pss_labels = dict_to_arr(pss_labels_dict)
pss_labels
```

```python
ttest = ttest_ind(stai_labels, pss_labels, nan_policy='raise', random_state=42)
```

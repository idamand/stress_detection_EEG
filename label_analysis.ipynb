```python
from scipy.stats import ttest_ind
from sklearn.metrics import adjusted_rand_score
import numpy as np
import utils.variables as var
from classifiers import k_means_clustering
from prepare_data import prepare_data_for_clustering
```

```python
data, stai_labels, pss_labels = prepare_data_for_clustering(data_type='raw', epoch_duration=299)
```

```python
cluster_labels = k_means_clustering(data)
cluster_labels
```

```python
adjusted_rand_score(labels_true=stai_labels, labels_pred=cluster_labels)
```

```python
adjusted_rand_score(labels_true=pss_labels, labels_pred=cluster_labels)
```

```python
ttest = ttest_ind(stai_labels, pss_labels, nan_policy='raise', random_state=42)
print('Results form t_test: \t', ttest)
```

```python
ttest_cluster_stai = ttest_ind(stai_labels, cluster_labels, nan_policy='raise', random_state=42)
print('Results form t_test: \t', ttest_cluster_stai)
```

```python
ttest_cluster_pss = ttest_ind(pss_labels, cluster_labels, nan_policy='raise', random_state=42)
print('Results form t_test: \t', ttest_cluster_pss)
```
